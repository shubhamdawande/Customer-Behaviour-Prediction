```python
import pandas as pd
pd.options.display.max_rows = 10
pd.options.display.max_columns = None

import json
```

```python
df = pd.read_csv('DataScience_CaseStudy_DataSet.csv')
```

```python
def parse_transaction_info(txns):
    txns = json.loads(txns.replace("\'", "\""))
    return txns
    
df['transaction_info_json'] = df['transaction_info'].apply(parse_transaction_info)
```

#### Transaction info column is in json format. First we parse json data and generate article, salesquantity and price columns. We will save the final dataframe for quick access.

```python
df_parsed = pd.DataFrame(columns=['csn', 'date', 'article', 'salesquantity', 'price'])

for x in range(len(df)):
    if x%1000==0:
        print(x)
        
    json_list = df.loc[x, 'transaction_info_json']
    for idx, item in enumerate(json_list):
        row = pd.DataFrame({
                          'csn': df.loc[x, 'csn'],
                          'date': df.loc[x, 'date'],
                          'article': item['article'], 
                          'salesquantity': item['salesquantity'], 
                          'price': item['price']
                          }, index=[x + idx])
        df_parsed = pd.concat([df_parsed, row])
        
df_parsed.head(10)
```

```python
df_parsed.to_csv('store_transactions.csv')
```
